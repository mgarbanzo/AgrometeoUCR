Contiene una forma interactiva de navegar los datos del Instituto Meteorológico Nacional disponibles desde el 2022-11-24 a las 10:00:00, hasta el día de hoy.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df = pd.read_csv("SantaCruz.csv")
```

```python
df.columns
```

```python
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
```

```python
for var in df.columns[1:]:
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df['TIMESTAMP'], y=df[var], mode='lines', name=var))

    fig.update_layout(
        title_text="Santa Cruz " + var,
        xaxis_title="Fecha y Hora",
        yaxis_title="Variable Medida",
    )

    fig.show()
```
